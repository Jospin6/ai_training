```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../data/home_prices/home_data.csv")
df
```

```python
df.columns
```

```python
data = df[["price", "sqft_living"]].rename(
    columns={"sqft_living": "area"}
)

print(data.head())
```

```python
print(data.to_string(index=False))
```

```python
model = LinearRegression()
model.fit(data[["area"]], data["price"])
print(model.coef_)
print(model.intercept_)
```
